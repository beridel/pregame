```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as MaxNLocator
from datetime import datetime, timedelta
import glob
import h5py
import numpy as np

mpl.rcParams['axes.xmargin'] = 0
```

```python
wv_list = sorted(glob.glob('../h5data/*h5'))

for flp in wv_list:
    with h5py.File(flp, 'r') as fp:
        amp = list(fp['amp'])
    x = np.linspace(0, len(amp), len(amp))
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(x,amp, linewidth=1, color='black')
    print(len(amp))
```
